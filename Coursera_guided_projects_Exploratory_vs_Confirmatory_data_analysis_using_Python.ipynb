```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
data=pd.read_excel('dataset.xlsx')
```

```python
data.dtypes
```

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
from_=data['Order Date'].min()
to_=data['Order Date'].max()
print("We have the sales info from",from_,'to',to_)
```

```python
data=data.sort_values(by='Order Date')
```

```python
data['Year']=pd.DatetimeIndex(data['Order Date']).year#extracts the year and returns the pandas series
```

```python
data['Month']=pd.DatetimeIndex(data['Order Date']).month
data['Day']=pd.DatetimeIndex(data['Order Date']).day
```

```python
data
```

```python
data_time_yearly_profit=data.groupby(['Year','Category']).agg({'Profit':'sum'}).reset_index()#pass the data and method of aggregation
data_time_yearly_profit
```

```python
px.line(data_time_yearly_profit,x='Year',y='Profit',color='Category')
```

```python
data_time_monthly_profit=data.groupby(['Year','Month','Category']).agg({'Profit':'sum'}).reset_index()
data_time_monthly_profit['Date']=data_time_monthly_profit.Year.astype(str)+'-'+data_time_monthly_profit.Month.astype(str)
```

```python
data_time_monthly_profit
```

```python
px.line(data_time_monthly_profit,x='Date',y='Profit',color='Category')
```

```python
#confirmatory data analysis is the process of using statistical summary and graphical representations
#to evaluate the validity of an assumption about the data at hand
```

```python
seasons={
    1:'Winter',
    2:'Spring',
    3:'Summer',
    4:'Fall'
}
```

```python
data['Season']=data.Month.astype(int)%12//3+1
data.Season=data.Season.map(seasons)
data.head()
```

```python
summer_data=data[data['Season']=='Summer']
summer_data.head()
```

```python
summer_data_agg=summer_data.groupby(['Year','Category','Season']).agg({'Quantity':'sum'}).reset_index()
```

```python
summer_data_agg
```

```python
px.bar(summer_data_agg,x=summer_data_agg['Year'].astype('str'),y='Quantity',
       color='Category',barmode='group')
```
